### Q1.1
Likelihood Function of the Linear Regression Model
The likelihood function for the linear regression model is given by:

$
p(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \Sigma) \propto \exp \left( -\frac{1}{2} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^\top \Sigma^{-1} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) \right)
$

where $\Sigma = \sigma^2 I$, indicating that the errors are independently and identically distributed normal with variance $\sigma^2$.

Prior Distribution:
\
Assume that $\boldsymbol{\beta}$ follows a multivariate normal prior:

$
p(\boldsymbol{\beta}) = \mathcal{MVN}(\boldsymbol{\beta}_0, \Sigma_{\beta})
$

which implies:

$
p(\boldsymbol{\beta}) \propto \exp \left( -\frac{1}{2} (\boldsymbol{\beta} - \boldsymbol{\beta}_0)^\top \Sigma_{\beta}^{-1} (\boldsymbol{\beta} - \boldsymbol{\beta}_0) \right)
$

Computing the Posterior Distribution Using Bayes' Rule
By Bayes' rule, the posterior distribution is:

$
p(\boldsymbol{\beta} \mid \mathbf{y}, \mathbf{X}, \Sigma) \propto p(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \Sigma) p(\boldsymbol{\beta})
$

Substituting the likelihood and prior:

$
p(\boldsymbol{\beta} \mid \mathbf{y}, \mathbf{X}, \Sigma) \propto \exp \left( -\frac{1}{2} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^\top \Sigma^{-1} (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) \right) \times \exp \left( -\frac{1}{2} (\boldsymbol{\beta} - \boldsymbol{\beta}_0)^\top \Sigma_{\beta}^{-1} (\boldsymbol{\beta} - \boldsymbol{\beta}_0) \right)
$

Rearranging the Exponent Terms
$
-\frac{1}{2} \left[ \boldsymbol{\beta}^\top \left( \mathbf{X}^\top \Sigma^{-1} \mathbf{X} + \Sigma_{\beta}^{-1} \right) \boldsymbol{\beta} - 2 \boldsymbol{\beta}^\top \left( \mathbf{X}^\top \Sigma^{-1} \mathbf{y} + \Sigma_{\beta}^{-1} \boldsymbol{\beta}_0 \right) \right]
$

Recognizing the quadratic form of a multivariate normal density, we identify the posterior covariance and mean:

$
\Sigma_{\beta \mid y} = \left( \mathbf{X}^\top \Sigma^{-1} \mathbf{X} + \Sigma_{\beta}^{-1} \right)^{-1}
$

$
\mu_{\beta \mid y} = \Sigma_{\beta \mid y} \left( \mathbf{X}^\top \Sigma^{-1} \mathbf{y} + \Sigma_{\beta}^{-1} \boldsymbol{\beta}_0 \right)
$

Thus, the posterior distribution of \$(\boldsymbol{\beta}\)$is:

$
\boldsymbol{\beta} \mid \mathbf{y}, \mathbf{X}, \Sigma \sim \mathcal{MVN} (\mu_{\beta \mid y}, \Sigma_{\beta \mid y})
$

This is the posterior distribution of $\boldsymbol{\beta}$given $\mathbf{y}$, $\mathbf{X}, $and $\Sigma$.

Model Assumptions

I analyze the posterior distribution of $ \sigma^2 $, the error variance in a linear regression model with design matrix $\ \mathbf{X} $, assuming an inverse-gamma prior.

#### Likelihood Function:
The likelihood of the observed data $\ \mathbf{y}$given parameters follows a multivariate normal distribution:

$
p(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \sigma^2) = \mathcal{MVN}(\mathbf{X} \boldsymbol{\beta}, \sigma^2 I)
$

which implies:

$
p(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \sigma^2) \propto (\sigma^2)^{-n/2} \exp \left( -\frac{(\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^\top (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})}{2\sigma^2} \right)
$

Prior for $\ \sigma^2 $:
I assume$  \sigma^2 $ follows an inverse-gamma prior distribution:

$
p(\sigma^2) = \text{Inv-Gamma} (\alpha^*, \beta^*)
$

which can be expressed as:

$
p(\sigma^2) \propto (\sigma^2)^{-(\alpha^*+1)} \exp \left( -\frac{\beta^*}{\sigma^2} \right)
$

Using Bayes' rule, the posterior distribution is proportional to the product of the likelihood and prior:

$
p(\sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta}) \propto p(\mathbf{y} \mid \mathbf{X}, \boldsymbol{\beta}, \sigma^2) p(\sigma^2)
$

Substituting the explicit expressions:

$
p(\sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta}) \propto (\sigma^2)^{-n/2} \exp \left( -\frac{\|\mathbf{y} - \mathbf{X} \boldsymbol{\beta}\|^2}{2\sigma^2} \right) \cdot (\sigma^2)^{-(\alpha^*+1)} \exp \left( -\frac{\beta^*}{\sigma^2} \right)
$

Rearranging terms:

$
p(\sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta}) \propto (\sigma^2)^{-(\alpha^* + n/2 + 1)} \exp \left( -\frac{\beta^* + \frac{1}{2} \|\mathbf{y} - \mathbf{X} \boldsymbol{\beta}\|^2}{\sigma^2} \right)
$

This follows the standard form of an inverse-gamma distribution:

$
\sigma^2 \mid \mathbf{y}, \mathbf{X}, \boldsymbol{\beta} \sim \text{Inv-Gamma} \left( \alpha^* + \frac{n}{2}, \beta^* + \frac{1}{2} \|\mathbf{y} - \mathbf{X} \boldsymbol{\beta}\|^2 \right)
$

Thus, the posterior remains inverse-gamma with updated parameters incorporating observed data.
